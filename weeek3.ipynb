```python

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 20)

y1 = 5*x**3 + 2*x - 1
y2 = -2*x*3 + x*2 + 100
y3 = 2*np.pi*x + 20

plt.plot(x, y1, label="5x^3 + 2x - 1")
plt.plot(x, y2, label="-2x^3 + x^2 + 100")
plt.plot(x, y3, label="2πx + 20")

plt.legend()
plt.grid()
plt.show()
```

```python

import matplotlib.pyplot as plt


x = [1, 2, 3]
y = [4, 5, 6]



plt.subplot(2,2,1)
plt.plot(x, y, marker='*', color='red')
plt.title("Star Marker")

plt.subplot(2,2,2)
plt.plot(x, y, marker='x', color='green')
plt.title("X Marker")

plt.subplot(2,2,3)
plt.plot(x, y, marker='^', color='blue')
plt.title("Triangle Marker")

plt.subplot(2,2,4)
plt.plot(x, y, marker='+', color='orange')
plt.title("Plus Marker")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt


days = [1, 2, 3, 4, 5]

# Distance covered
enfield = [50, 40, 70, 80, 20]
honda = [80, 20, 20, 50, 60]
yamaha = [70, 20, 60, 40, 60]
ktm = [40, 20, 30, 70, 60]

plt.figure(figsize=(10,8))

# Subplot 1
plt.subplot(2,2,1)
plt.plot(days, enfield, marker='o')
plt.title("Enfield")
plt.xlabel("Days")
plt.ylabel("Distance Covered")
plt.grid()

# Subplot 2
plt.subplot(2,2,2)
plt.plot(days, honda, marker='*')
plt.title("Honda")
plt.xlabel("Days")
plt.ylabel("Distance Covered")
plt.grid()

# Subplot 3
plt.subplot(2,2,3)
plt.plot(days, yamaha, marker='x')
plt.title("Yamaha")
plt.xlabel("Days")
plt.ylabel("Distance Covered")
plt.grid()

# Subplot 4
plt.subplot(2,2,4)
plt.plot(days, ktm, marker='^')
plt.title("KTM")
plt.xlabel("Days")
plt.ylabel("Distance Covered")
plt.grid()

plt.suptitle("Bike details in Plot")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3,3,100)

# Parabola
y1 = x**2

# Circle
y2 = np.sqrt(9 - x**2)
y3 = -np.sqrt(9 - x**2)

plt.plot(x, y1, label="Parabola")
plt.plot(x, y2, label="Circle")
plt.plot(x, y3)

plt.title("Intersection of Two Conic Sections")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = 2
b = 2
r = 4

theta = 0
x = []
y = []

while theta <= 2*np.pi:
    x.append(a + r*np.cos(theta))
    y.append(b + r*np.sin(theta))
    theta += 0.01

plt.plot(x, y)

plt.title("Circle with Center (2,2) and Radius 4")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()
plt.axis('equal')

plt.show()
```
