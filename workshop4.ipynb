```python
import matplotlib.pyplot as plt

# Data
colors = ["Red", "Silver", "Blue", "White", "Green"]
values = [123, 230, 78, 193, 12]

# ---------------- a) Bar Chart ----------------
plt.figure()
plt.bar(colors, values)
plt.title("Car Color Distribution (Bar Chart)")
plt.xlabel("Car Color")
plt.ylabel("Number Seen")
plt.show()

# ---------------- b) Bar width = 0.3 ----------------
plt.figure()
plt.bar(colors, values, width=0.3)
plt.title("Bar Chart (Width = 0.3)")
plt.show()

# ---------------- c) Horizontal Bar Chart ----------------
plt.figure()
plt.barh(colors, values)
plt.title("Horizontal Bar Chart")
plt.show()

# ---------------- d) Bar colors same as car colors ----------------
mycolors = ["red", "silver", "blue", "lightgray", "green"]

plt.figure()
plt.bar(colors, values, color=mycolors)
plt.title("Bar Chart with Matching Colors")
plt.show()

# ---------------- e) Pie Chart with Title ----------------
plt.figure()
plt.pie(values, labels=colors)
plt.title("Car Color Distribution (Pie Chart)")
plt.show()

# ---------------- f) Pie Chart Variations ----------------

# A) Explode
plt.figure()
myexplode = [0.3, 0, 0, 0, 0]
plt.pie(values, labels=colors, explode=myexplode)
plt.title("Pie Chart with Explode")
plt.show()

# B) Explode + Shadow
plt.figure()
plt.pie(values, labels=colors, explode=myexplode, shadow=True)
plt.title("Pie Chart with Explode + Shadow")
plt.show()

# C) Legend with Title
plt.figure()
plt.pie(values)
plt.legend(labels=colors, title="Car Colours:")
plt.title("Pie Chart with Legend")
plt.show()

# D) Custom Colors
plt.figure()
mycolours = ["magenta", "green", "cyan", "red", "yellow"]
plt.pie(values, labels=colors, colors=mycolours)
plt.title("Pie Chart with Custom Colors")
plt.show()
```
